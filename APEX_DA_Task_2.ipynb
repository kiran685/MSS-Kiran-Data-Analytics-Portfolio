```python
import pandas as pd

# Load the laptop_scrap_data.csv file
df_laptops = pd.read_csv('/content/laptop_scrap_data.csv', encoding='latin1', sep=None, engine='python', on_bad_lines='skip')
print("First 5 rows of the dataset:")
display(df_laptops.head())
```

```python
print("\nDataFrame Information:")
df_laptops.info()
```

```python
print("\nDescriptive Statistics:")
display(df_laptops.describe(include='all'))
```

```python
print('\nShape before dropping missing values:', df_laptops.shape)
# Drop rows with any missing values
df_laptops.dropna(inplace=True)
print('Shape after dropping missing values:', df_laptops.shape)

print('\nDataFrame Information after dropping missing values:')
df_laptops.info()
```

```python
print('\nUnique values for TouchScreen:\n', df_laptops['TouchScreen'].value_counts())
print('\nUnique values for Ips:\n', df_laptops['Ips'].value_counts())
print('\nUnique values for Dedicated_Gpu:\n', df_laptops['Dedicated_Gpu'].value_counts())
print('\nUnique values for OpSys:\n', df_laptops['OpSys'].value_counts())
print('\nUnique values for Storage_Type:\n', df_laptops['Storage_Type'].value_counts())
```

```python
print('Shape before removing "No Storage Data" from Storage_Type:', df_laptops.shape)
# Remove rows where 'Storage_Type' is 'No Storage Data'
df_laptops = df_laptops[df_laptops['Storage_Type'] != 'No Storage Data']
print('Shape after removing "No Storage Data" from Storage_Type:', df_laptops.shape)

print('\nUpdated unique values for Storage_Type:\n', df_laptops['Storage_Type'].value_counts())
```

```python
print('\nDataFrame Information after final cleaning:')
df_laptops.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df_laptops['Price'], bins=50, kde=True)
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(x='Company', y='Price', data=df_laptops)
plt.title('Laptop Price Distribution by Company')
plt.xlabel('Company')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Price Distribution by RAM (GB)

Let's analyze the impact of RAM on laptop prices. We'll use a box plot to see how prices vary with different amounts of installed RAM.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Ram_GB', y='Price', data=df_laptops.sort_values('Ram_GB'))
plt.title('Laptop Price Distribution by RAM (GB)')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Correlation Heatmap of Numerical Features

To understand the relationships between different numerical variables, including their correlation with 'Price', we'll generate a correlation heatmap.

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df_laptops.select_dtypes(include=['float64']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

### Business Questions & Simulated SQL Queries (using Pandas)

Here are some business questions to explore the `df_laptops` dataset. We'll answer them using pandas, simulating SQL-like operations like filtering and aggregation.

#### Question 1: What are the top 5 laptop companies by average price?

```python
top_companies_by_price = df_laptops.groupby('Company')['Price'].mean().nlargest(5).reset_index()
print("Top 5 laptop companies by average price:")
display(top_companies_by_price)
```

#### Question 2: What is the distribution of laptops by 'TypeName' (e.g., Gaming, Ultrabook) and their average price?

```python
type_name_distribution = df_laptops.groupby('TypeName')['Price'].agg(['count', 'mean']).reset_index()
type_name_distribution.rename(columns={'count': 'Number_of_Laptops', 'mean': 'Average_Price'}, inplace=True)
print("Distribution of laptops by TypeName and their average price:")
display(type_name_distribution.sort_values('Number_of_Laptops', ascending=False))
```

#### Question 3: How does the average price differ between 'SSD Only' and 'HDD Only' storage types?

```python
storage_type_prices = df_laptops.groupby('Storage_Type')['Price'].mean().reset_index()
print("Average price by storage type:")
display(storage_type_prices)
```

#### Question 4: Which operating system (OpSys) has the highest average price for laptops with more than 8GB RAM?

```python
high_ram_laptops = df_laptops[df_laptops['Ram_GB'] > 8]
avg_price_by_os_high_ram = high_ram_laptops.groupby('OpSys')['Price'].mean().nlargest(1).reset_index()
print("Operating system with the highest average price for laptops with >8GB RAM:")
display(avg_price_by_os_high_ram)
```

#### Question 5: What is the average price of laptops that have both a TouchScreen and an IPS display?

```python
touch_ips_laptops = df_laptops[(df_laptops['TouchScreen'] == 1) & (df_laptops['Ips'] == 1)]
avg_price_touch_ips = touch_ips_laptops['Price'].mean()
print(f"Average price of laptops with both TouchScreen and IPS display: {avg_price_touch_ips:.2f}")
```
