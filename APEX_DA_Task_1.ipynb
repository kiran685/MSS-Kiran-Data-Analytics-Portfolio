### Data Quality Assessment and Cleaning

Based on our initial profiling, we have identified a few areas that require attention. Let's systematically check for and address them:

1.  **Duplicate Rows:** Ensure there are no completely duplicated entries in the dataset.
2.  **`addiction_score`:** Negative values were observed. We need to investigate if these are errors or a specific scoring methodology. For now, we will identify and count them.
3.  **`grades`:** Values over 100 were observed. We need to investigate if these are valid (e.g., bonus points) or errors. For now, we will identify and count them.
4.  **Categorical Columns Consistency:** Check unique values in `gender`, `gaming_genre`, and `stress_level` to ensure consistency in spelling and formatting.

```python
# 1. Check for Duplicate Rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

if duplicate_rows > 0:
    print("\nDropping duplicate rows...")
    df.drop_duplicates(inplace=True)
    print("Duplicate rows dropped.")
    print(f"New DataFrame shape: {df.shape}")
else:
    print("No duplicate rows found.")
```

```python
# 2. Investigate `addiction_score` for negative values
negative_addiction_scores = df[df['addiction_score'] < 0]
print(f"\nNumber of entries with negative 'addiction_score': {len(negative_addiction_scores)}")
if not negative_addiction_scores.empty:
    display(negative_addiction_scores.head())
    print("\nThese negative values need further investigation. For now, we will treat them as invalid and cap them at 0, assuming addiction scores cannot be negative.")
    df['addiction_score'] = df['addiction_score'].apply(lambda x: max(x, 0))
    print("Negative 'addiction_score' values have been capped at 0.")
else:
    print("No negative 'addiction_score' values found.")
```

```python
# 3. Investigate `grades` for values greater than 100
high_grades = df[df['grades'] > 100]
print(f"\nNumber of entries with 'grades' > 100: {len(high_grades)}")
if not high_grades.empty:
    display(high_grades.head())
    print("\nGrades greater than 100 might indicate a bonus system or data entry errors. For now, we will cap them at 100, assuming a standard grading system.")
    df['grades'] = df['grades'].apply(lambda x: min(x, 100))
    print("'grades' values greater than 100 have been capped at 100.")
else:
    print("No 'grades' values greater than 100 found.")
```

```python
# 4. Check consistency of categorical columns
print("\nUnique values in 'gender':")
print(df['gender'].unique())

print("\nUnique values in 'gaming_genre':")
print(df['gaming_genre'].unique())

print("\nUnique values in 'stress_level':")
print(df['stress_level'].unique())

# If inconsistencies are found (e.g., 'Male', 'male'), they would be handled here.
# For example:
# df['gender'] = df['gender'].str.capitalize()
# df['stress_level'] = df['stress_level'].replace({'low': 'Low', 'high': 'High'})
# Based on the output of .unique(), we can add specific cleaning steps if needed.
```

```python
# Output the final, analysis-ready dataset to a CSV file
df.to_csv('Gaming_Academic_Performance_cleaned.csv', index=False)
print("Cleaned dataset saved as 'Gaming_Academic_Performance_cleaned.csv'")
```
